```python
import numpy as np

x_0 = 0
y = 1
h = 0.1
xs = []
ys = []

def f(x,y):
    return x**2 - y

for i in np.arange(x_0, 0.6, h):
    y = f(i,y) * h + y
    xs.append(0.1 + i)
    ys.append(y)
    print(i,y)
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0,1,100)

y = x**2 - 2*x - np.exp(-x) + 2


plt.scatter(xs,ys)
plt.plot(x,y)
```
