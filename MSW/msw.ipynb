```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
# 6 grafu

plt.figure(figsize=(8, 4))

plt.subplot(2, 3, 1)
x = np.linspace(0, 2*np.pi)
plt.plot(x, np.sin(x))

plt.subplot(2, 3, 2)
x = np.linspace(0, 2*np.pi)
plt.plot(x, np.cos(x))

plt.subplot(2,3,3)
x = np.linspace(0, 2*np.pi)
plt.plot(x, np.tan(x))

plt.subplot(2, 3, 4)
x = np.linspace(0, 2*np.pi)
plt.scatter(x, np.sin(x))

plt.subplot(2, 3, 5)
x = np.linspace(0, 2*np.pi)
plt.scatter(x, np.cos(x))

plt.subplot(2,3,6)
x = np.linspace(0, 2*np.pi)
plt.scatter(x, np.tan(x))


plt.show()
```
