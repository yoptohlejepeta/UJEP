```python
# spanish system
import numpy as np

def pois(lambda_param, k):
    x_values = np.arange(0, k)  # Values from 0 to 19 for the x-axis
    a = np.exp(-lambda_param)
    b = np.power(lambda_param, x_values)
    c = np.array([np.math.factorial(x) for x in x_values])
    poisson_values = (a * b) / c

    return poisson_values[0]


a = [0,0,0,0,1]
p = pois(0.1, len(a))
print(p)


matice_p = [[p, 0,0,0, 1-p],
            [p, 0,0,0, 1-p],
            [0, p,0,0, 1-p],
            [0, 0,p,0, 1-p],
            [0, 0,0,p, 1-p]]



while True:
    a_new = np.dot(a, matice_p)
    if np.allclose(a, a_new, rtol=1e-20):
        break
    a = a_new
    
print(a_new)
print(sum(a_new))
```
