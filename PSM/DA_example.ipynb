```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn import datasets
```

```python
# Načtení datasetu Iris
iris = datasets.load_iris()
# Definice vysvětlujících proměnných
X = iris.data
# Definice cílové proměnné
y = iris.target
```

```python
print(iris.target_names)
```

```python
# Vytvoření instance modelu kanonické diskriminační analýzy
lda = LinearDiscriminantAnalysis(n_components=2)
```

```python
# Trénování modelu na datech
X_lda = lda.fit(X, y).transform(X)
```

```python
# Vykreslení výsledků
import matplotlib.pyplot as plt

colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], iris.target_names):
    plt.scatter(X_lda[y == i, 0], X_lda[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
    
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of IRIS dataset')
plt.show()
```

```python
# Kvadratická diskriminační analýza příklad implementace na python
# Import knihoven
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.model_selection import train_test_split
```

```python
# Načtení datasetu iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
 #Rozdělení dat na trénovací a testovací sady
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Vytvoření instance modelu kvadratické diskriminační analýzy
qda_model = QuadraticDiscriminantAnalysis()
```

```python
# Trénování modelu
qda_model.fit(X_train, y_train)
```

```python
# Předpověď na testovacích datech
predicted_classes = qda_model.predict(X_test)
```

```python
# Výpis přesnosti predikce
accuracy = qda_model.score(X_test, y_test)
print("Přesnost predikce: ", accuracy)
```

```python
# k-NN method
# nahrání knihoven
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
```

```python
# nahrání iris dat
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# výběr trénovacích dat a testovacích dat
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# konfuzní matice
confusion_matrix(y_test, y_pred)
```

```python
# CART
# nahrání knihoven
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn import tree
import graphviz
```

```python
# nahrání iris dat
iris = load_iris()
X = iris.data
y = iris.target

# výběr trénovacích dat a testovacích dat
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# trénování modelu
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
# předpověď na testovacích datech
y_pred = dt.predict(X_test)
```

```python
# konfuzní matice
confusion_matrix(y_test, y_pred)
```

```python
# vizuální zobrazení rozhodovacího stromu
dot_data = tree.export_graphviz(dt, out_file=None, 
                      feature_names=iris.feature_names,  
                      class_names=iris.target_names,  
                      filled=True, rounded=True,  
                      special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

```python
# Random Forest
# nahrání knihoven
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

```python
# nahrání iris dat
iris = load_iris()
X = iris.data
y = iris.target

# výběr trénovacích dat a testovacích dat
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# trénování modelu
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)
```

```python
# předpověď na testovacích datech
y_pred = rf.predict(X_test)
```

```python
# konfuzní matice
confusion_matrix(y_test, y_pred)
```

```python
# vizuální zobrazení důležitosti charakteristik
features = iris.feature_names
importances = rf.feature_importances_
indices = np.argsort(importances)
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
