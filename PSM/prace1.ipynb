```python
import pandas as pd
import plotly.express as px
from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_excel("data/Concrete_Data.xls", sheet_name="Sheet1")
data.head()
```

```python
data.columns = ["cement", "blast_furncace_slag", "fly_ash", "water", "superplasticizer", "coarse_aggregate", "fine_aggregate", "age", "concrete_compressive_strength"]
data.head()
```

```python
data.corr()
```

```python
fig = px.imshow(data.corr())
fig.show()
```

```python
fig = px.scatter_matrix(data)
fig.show()
```

```python
x = data.iloc[:,0:8].values
y = data["concrete_compressive_strength"]
```

```python
fig = px.box(x)
fig.show()
```

```python
min_max_scaler = preprocessing.StandardScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)

fig = px.box(df)
fig.show()
```

```python
pca = PCA(n_components=4)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

x_pca_train = pca.fit_transform(X_train)
x_pca_test = pca.fit_transform(X_test)
```

```python
model1 = LinearRegression()
model1.fit(x_pca_train, y_train)
predict = model1.predict(x_pca_test)

print(model1.score(x_pca_test, y_test))
print(np.sqrt(mean_squared_error(y_test, predict)))
print(model1.coef_)
print(model1.intercept_)
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)

x_poly = poly.fit_transform(x_pca_train)

model2 = LinearRegression()

model2.fit(x_poly, y_train)
predict = model2.predict(x_pca_test)

print(np.sqrt(mean_squared_error(y_test, predict)))
print(model1.coef_)
print(model1.intercept_)
```
