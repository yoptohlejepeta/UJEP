```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy.random as random

X, y = make_blobs(n_samples=1000, centers=2, n_features=2, random_state=1)

plt.scatter(X[:,0], X[:,1])
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def J(w, data, labels):
    cleny = []
    for i,j in zip(data, labels):
        cleny.append((j - sigmoid(np.dot(w, i)))**2)
    return sum(cleny)
```

```python
w = random.rand(2)
alpha = 0.1
count = 1

for i in range(50):
    if J(w - alpha * sigmoid(w)*(1-sigmoid(w)), X, y) < J(w, X, y):
        count += 1
        w = w - alpha * sigmoid(w)*(1-sigmoid(w))
    else:
        alpha = alpha * 0.9


```

```python
x_line = np.linspace(np.min(X[:,0]),np.max(X[:,0]), 100)
y_line = w[0]*x_line + w[1]

plt.scatter(X[:,0], X[:,1])
plt.plot(x_line, y_line)
```
