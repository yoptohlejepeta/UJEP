```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(C=1, penalty="l2", solver="liblinear")
X, y = make_blobs(n_samples=1000, centers=2, n_features=2, random_state=0)
model.fit(X, y)
y_hat = model.predict(X)

predpovedi = 0
for trida_predpoved, realita in zip(y_hat, y):
    # print(f"{trida_predpoved - realita}|", end="")
    predpoved = trida_predpoved - realita
    if predpoved != 0:
        predpovedi += 1

barvy1 = []
for item in y_hat:
    if item == 0:
        barvy1.append("red")
    else:
        barvy1.append("green")

barvy2 = []
for item in y:
    if item == 0:
        barvy2.append("red")
    else:
        barvy2.append("green")

print(predpovedi)
plt.scatter(X[:, 0], X[:, 1], color=barvy1)
plt.show()

plt.scatter(X[:, 0], X[:, 1], color=barvy2)
plt.show()
```
